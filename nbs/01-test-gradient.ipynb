#### Test: Gradient image

```python
import numpy as np
from PIL import Image

# Create a gradient image
gradient = np.linspace(0, 255, 500).astype(np.uint8)
gradient = np.tile(gradient, (500, 1))
Image.fromarray(gradient).save('test_gradient.png')

# Display it to verify
Image.open('test_gradient.png')
```

```python
import subprocess
import os

# Get the project root (parent of nbs/)
project_root = os.path.dirname(os.getcwd())

result = subprocess.run(
    ['python', os.path.join(project_root, 'src/main.py'), 
    '--input', 'test_gradient.png', 
    '--output', 'test_output.png', 
    '--no-quote'],
    capture_output=True,
    text=True
)

print("Return code:", result.returncode)
print(result.stdout)
if result.stderr:
    print("Errors:", result.stderr)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load both images
original = Image.open('test_gradient.png')
processed = Image.open('test_output.png')

# Display side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.imshow(original, cmap='gray')
ax1.set_title('Original Gradient')
ax1.axis('off')

ax2.imshow(processed, cmap='gray')
ax2.set_title('Processed Output')
ax2.axis('off')

plt.tight_layout()
plt.show()
```
